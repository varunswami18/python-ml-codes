## Concat

```python
india_weather = pd.DataFrame({
    "city": ["mumbai", "delhi", "banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 70]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city":["new york", "chicago", "orlando"],
    "temperature":[21, 14,35],
    "humidity":[68, 65, 75]
})
us_weather
```

```python
# concat
df = pd.concat([india_weather, us_weather], ignore_index=True)
df
```

```python
# using keys in concatenation
df = pd.concat([india_weather, us_weather], keys=['India', 'US']) # can not use ignore_index
df
```

```python
# accessing dataframe using key
df.loc['US']
```

```python
# concatenating horizontally
df = pd.concat([india_weather, us_weather],axis=1, ignore_index=True)
df
```

```python
# aligning same type of columns using index
temp_df = pd.DataFrame({
    "city": ["mumbai", "delhi"],
    "temperature": [32,45]
})

humid_df = pd.DataFrame({
    "city": ["delhi", "banglore", 'mumbai'],
    "humidity": [80, 60, 70]
})

print(temp_df, "\n")
print(humid_df, "\n")

df = pd.concat([temp_df, humid_df], axis=1)
print(df, "\n")

temp_df = pd.DataFrame({
    "city": ["mumbai", "delhi"],
    "temperature": [32,45]
}, index=[0, 1])

humid_df = pd.DataFrame({
    "city": ["delhi", "banglore", 'mumbai'],
    "humidity": [80, 60, 70]
}, index=[1,2,0])

print(temp_df, "\n")
print(humid_df,"\n")

df = pd.concat([temp_df, humid_df], axis=1)
print(df)
```

```python
# reomoving same name columns
cols = []
count = 1
for column in df.columns:
    if column == 'city':
        cols.append(f'city_{count}') #renaming same name columns with different names
        count+=1
        continue
    cols.append(column)
df.columns = cols #adding column list in dataframe
print(df) 
df.drop('city_1', inplace=True, axis=1) #droppping column that is not needed
df.rename({'city_2':'city'}, axis=1, inplace=True) #renaming again
city=df.pop('city') #to move the column to desired position- pop, store, add in desired location
df.insert(0, 'city', city)
```

## Crosstab

#### In statistics, a contingency table(also known as a cross tabulation or crosstab) is a type of matrix fromat that displays(multivariate) freuency distribution of the variables

```python
df = pd.read_excel("D:/thedesktop/py-master/pandas/13_crosstab/survey.xls")
df
```

```python
# crosstab
pd.crosstab(df.Sex, df.Handedness, margins=True)
```

```python
# multiple variables
pd.crosstab(df.Nationality,[df.Sex, df.Handedness], margins=True)
```

```python
pd.crosstab([df.Nationality, df.Sex], df.Handedness, margins=True)
# Shift Tab tells about functions
```

```python
# aggregate functions with crosstab
pd.crosstab([df.Sex], [df.Handedness], values=df.Age, aggfunc=np.average) #values are those which are to be investigated
```

```python
# Create your own Holiday Calendar
from pandas.tseries.holiday import AbstractHolidayCalendar, nearest_workday, Holiday

class myBirthDayCalendar(AbstractHolidayCalendar):
    rules=[
        Holiday("My Birthday", month=2, day=25)
    ]
myCal = CustomBusinessDay(calendar=myBirthDayCalendar())
myCal
```
