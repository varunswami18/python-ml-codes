# Seaborn
Seaborn is a Python library for data visualization build on top of the matplotlib Python library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df
```

## Distribution plot

```python
sns.distplot(crash_df['not_distracted'])
```

```python
sns.distplot(crash_df['not_distracted'], kde=False)
```

```python
sns.distplot(crash_df['not_distracted'], bins=20)
```

## Joint Plot

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
#kernel density estimation
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

## KDE Plot

```python
sns.kdeplot(crash_df['alcohol'])
```

## Pair Plots

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
sns.pairplot(tips_df, hue='sex', palette=['blue', 'magenta'])
```

## Rug Plot

```python
sns.rugplot(tips_df['tip']) #denser at more populated points
```

## Styling

```python
sns.set_style('darkgrid')
plt.figure(figsize=(8, 4))
sns.set_context('paper', font_scale=1.4)
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(8, 4))
sns.set_context('talk', font_scale=1.4)
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('darkgrid')
sns.set_context('paper')
plt.figure(figsize=(8, 4))
sns.set_context('poster', font_scale=1.4)
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
sns.despine(left=True, bottom=True)#removes axis line
```

# Categorical Plots

## Bar Plots

```python
sns.set_context('paper', font_scale=1.4)
sns.barplot(x='sex', y='total_bill', data=tips_df)
```

## Count Plot

```python
#counts the occurences
sns.countplot(x='sex', data=tips_df)
```

## Boxplot

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

## Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_df, hue='sex', split=True)
```

## Strip Plot

```python
plt.figure(figsize=(8, 5))
sns.stripplot(x='day', y='total_bill', data=tips_df)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True) #jitter spread out data
```

```python
sns.stripplot(x='day', y='total_bill', data=tips_df, jitter=True, hue='sex', dodge=True)#dodge separates data
```

## Swarm Plot

```python
#combination of violin and strip
sns.violinplot(x='day', y='total_bill', data=tips_df)
sns.swarmplot(x='day', y='total_bill', data=tips_df, color='white')
```

# Palletes

```python
plt.figure(figsize=(8, 5))
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot(x='day', y='total_bill', data=tips_df, hue='sex', palette='magma')
```

# Matrix Plots

## Heatmaps

```python
plt.figure(figsize=(8, 6))
sns.set_context('paper', font_scale=1.4)

crash_mx = crash_df.corr()
sns.heatmap(crash_mx, annot=True, cmap='Blues')
plt.show()
```

```python
flights = sns.load_dataset('flights')
flights = flights.pivot_table(index='month', columns='year', values='passengers')
flights
```

```python
sns.heatmap(flights, cmap='Blues', linecolor='white', linewidth=1)
```

# Cluster Map

```python
iris = sns.load_dataset('iris')
species = iris.pop('species')
sns.clustermap(iris)
```

# Pair Grid

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue="species")
iris_g.map(plt.scatter)
```

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue="species")
iris_g.map_diag(plt.hist) # in dioagonal
iris_g.map_offdiag(plt.scatter) #not in diagonal
```

# Facet Frid

```python
tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')
tips_fg.map(plt.hist, 'total_bill', bins=8, )
```

```python
kws = dict(s=50, linewidth=0.5, edgecolor='w')
tips_fg = sns.FacetGrid(tips_df, col='sex', hue='smoker', height=4, aspect=1.3,
                       hue_order=['Yes', 'No'],
                       hue_kws=dict(marker=['^', 'v']))
tips_fg.map(plt.scatter, 'total_bill', 'tip', **kws)
```

```python
att_df = sns.load_dataset('attention')
att_fg = sns.FacetGrid(att_df, col='subject', col_wrap=5, height=1.5)
att_fg.map(plt.plot, 'solutions', 'score', marker='.')
```

# Regression Plots

```python
plt.figure(figsize=(8, 6))
sns.set_context('paper', font_scale=1.4)
sns.lmplot(x='total_bill', y='tip', hue='sex', data=tips_df, markers=['o', '^'],
          scatter_kws={'s':100, 'linewidth':0.5})
```
