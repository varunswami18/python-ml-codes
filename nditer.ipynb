```python
import numpy as np
```

```python
a = np.arange(12).reshape(3, 4)
print("Array is:\n", a)
```

```python
# nditer
print("C order array printing:")
for x in np.nditer(a, order='C'): # C order
    print(x)
print("\n")
print("Fortran order array prinitng:")
for x in np.nditer(a, order='F'): # Fortran order
    print(x)
```

```python
# updating arrays
for x in np.nditer(a, op_flags=['readwrite']):
    x[...]=x*x
print("Sqaure of Array:\n", a)
```

```python
# workin with two arrays simultaneously
b=np.arange(3, 15, 4).reshape(3, 1)
print("Array a is:\n", a)
print("\nArray b is:\n", b)
print("\nWorking with arrays simultaneously")
for x, y in np.nditer([a, b]):
    print(x, y)
```
